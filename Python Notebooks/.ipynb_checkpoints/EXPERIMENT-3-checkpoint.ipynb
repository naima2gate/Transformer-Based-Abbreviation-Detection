## Imports

```python
import nltk
import wandb
import string
import random
import warnings
import multiprocessing
import seaborn as sns
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras.layers import Flatten
from datasets import load_dataset
from sklearn.metrics import confusion_matrix, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Embedding, 
    Conv1D, 
    GlobalMaxPooling1D, 
    Dense, 
    TimeDistributed,
    LSTM,
    Bidirectional
)
from tensorflow.keras.preprocessing.sequence import pad_sequences
from gensim.models import (
    Word2Vec,
    FastText
)
from tqdm.notebook import tqdm
from collections import Counter
from nltk.corpus import stopwords
from seqeval.metrics import classification_report as seqeval_classification_report

# Ignore FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

## Load the Dataset

```python
datasets = load_dataset("surrey-nlp/PLOD-CW")
print("Dataset Loaded.")
```

## Preprocessing the Dataset

```python
stop_words = set(stopwords.words('english'))
def preprocess(dataset):
    tokens = []
    pos_tags = []
    ner_tags = []
    for token, pos_tag, ner_tag in zip(dataset['tokens'], dataset['pos_tags'], dataset['ner_tags']):
        if token.lower() not in stop_words and token not in string.punctuation:
            tokens.append(token.lower())
            pos_tags.append(pos_tag)
            ner_tags.append(ner_tag)
    dataset['tokens'] = tokens
    dataset['pos_tags'] = pos_tags
    dataset['ner_tags'] = ner_tags
    return dataset
    
# Apply preprocessing
datasets = datasets.map(preprocess)
print("Preprocessing Successful.")
```

## Create Label Mapping

```python
label_list = ['B-O', 'B-AC', 'I-LF', 'B-LF']
# Create label to index mapping
label_to_index = {label: idx for idx, label in enumerate(label_list)}
# Create index to label mapping
index_to_label = {idx: label for label, idx in label_to_index.items()}
```

## Dataset Preparation with Word Embeddings

```python
num_classes = len(label_list)

def prepare_dataset(dataset, embeddings):
    unk_index = embeddings.wv.key_to_index[UNK_TOKEN]  # Get index for UNK token
    sequences = [
        [embeddings.wv.key_to_index.get(word, unk_index) for word in tokens]
        for tokens in dataset['tokens']
    ]
    max_len = max(len(s) for s in sequences)
    X = pad_sequences(sequences, maxlen=max_len, padding='post')

    # Transform labels using label mapping
    y = pad_sequences([[label_to_index[tag] for tag in tags] for tags in dataset['ner_tags']], maxlen=max_len, padding='post')
    y = np.array([to_categorical(yt, num_classes=num_classes) for yt in y])
    
    return np.array(X), y
```

## BiLSTM Model Architecture

```python
class BiLSTMModel(tf.keras.Model):
    def __init__(self, embeddings):
        super(BiLSTMModel, self).__init__()
        self.embedding = tf.keras.layers.Embedding(input_dim=len(embeddings.wv.key_to_index) + 1,
                                                    output_dim=embeddings.vector_size, trainable=True)
        self.lstm = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=100, return_sequences=True, dropout=0.2, recurrent_dropout=0.2))
        self.time_distributed_1 = tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(100, activation='relu'))
        self.dropout = tf.keras.layers.Dropout(0.2)
        self.time_distributed_2 = tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(num_classes, activation='softmax'))

    def call(self, inputs):
        x = self.embedding(inputs)
        x = self.lstm(x)
        x = self.time_distributed_1(x)
        x = self.dropout(x)
        return self.time_distributed_2(x)
```

# SYSTEM 1: Word2Vec Embeddings

## Training a Word2Vec Model

```python
# Training the Word2Vec model
num_features = 100
min_word_count = 1
num_workers = multiprocessing.cpu_count()
num_workers = 2
window_size = 3
subsampling = 1e-3

# Add a special token for unknown words
UNK_TOKEN = "<UNK>"
all_tokens = [tokens + [UNK_TOKEN] for tokens in datasets['train']['tokens']]

# Train Word2Vec including the UNK token
w2v_model = Word2Vec(all_tokens, workers=num_workers, vector_size=num_features, min_count=min_word_count, window=window_size, sample=subsampling)

print('Word2Vec model trained.')
```

```python
model_name = "my_word2vec_model"
w2v_model.save(model_name)
#from gensim.models.word2vec import Word2Vec
#model_name = "my_word2vec_model"
w2v_embeddings = Word2Vec.load(model_name)
vocab_word2vec = w2v_embeddings.wv.index_to_key
```

## Model Definition and Compilation

```python
# Create an instance of the BiLSTMModel
model_w2v = BiLSTMModel(w2v_embeddings)

# Compile the model
model_w2v.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

## Dataset Preparation

```python
# Prepare datasets
X_train, y_train = prepare_dataset(datasets['train'], w2v_embeddings)
X_val, y_val = prepare_dataset(datasets['validation'], w2v_embeddings)
X_test, y_test = prepare_dataset(datasets['test'], w2v_embeddings)
```

## Training the Model

```python
history = model_w2v.fit(X_train, y_train, epochs=5, batch_size=5, validation_data=(X_val, y_val))
print('Model trained.')
```

## Predictions on Test Set

```python
max_len = 0
for sequence in datasets['test']['tokens']:
    if len(sequence) > max_len:
        max_len = len(sequence)
max_len
```

```python
# Make predictions on the test data
y_pred = model_w2v.predict(X_test)

# Convert predictions and ground truth labels to categorical
y_pred_classes = np.argmax(y_pred, axis=2)
y_test_classes = np.argmax(y_test, axis=2)

# Flatten the arrays to compute metrics for each class
y_pred_flat = y_pred_classes.flatten() 
y_test_flat = y_test_classes.flatten()  

# Convert flat arrays to lists of labels for each sequence
y_test_labels = [index_to_label[label_id] for label_id in y_test_flat]
y_pred_labels = [index_to_label[label_id] for label_id in y_pred_flat]

# Reshape these lists to match the original number of samples and their sequence lengths
y_test_labels = [y_test_labels[i:i + max_len] for i in range(0, len(y_test_flat), max_len)]
y_pred_labels = [y_pred_labels[i:i + max_len] for i in range(0, len(y_pred_flat), max_len)]
```

## Generating Classification Report

```python
# Compute classification report
report = seqeval_classification_report(y_test_labels, y_pred_labels)

# Print classification report
print("Classification Report:")
print(report)
```

## Error Analysis

### Plotting Confusion Matrix

```python
def plot_cm(y_true, y_pred, figsize=(10,10)):
    # Flatten the lists of lists to get a single list of labels
    y_true_flat = [label for sublist in y_true for label in sublist]
    y_pred_flat = [label for sublist in y_pred for label in sublist]

    # Compute confusion matrix
    cm = confusion_matrix(y_true_flat, y_pred_flat, labels=np.unique(y_true_flat))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    
    # Generate annotations for each cell
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = '0'
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
                
    # Create DataFrame for heatmap
    cm = pd.DataFrame(cm, index=np.unique(y_true_flat), columns=np.unique(y_true_flat))
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    
    # Plot heatmap
    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(cm, cmap="GnBu", annot=annot, fmt='', ax=ax)

    # Annotate each cell
    for i in range(nrows):
        for j in range(ncols):
            plt.text(j + 0.5, i + 0.5, annot[i, j], ha='center', va='center', color='black')

    plt.show()

plot_cm(y_test_labels, y_pred_labels)
```

### Plotting Error Distribution

```python
# Function to plot error distribution
def plot_error_distribution(true_labels_flat, true_predictions_flat, label_list):
    # Flatten the lists of lists to get a single list of labels
    y_true_flat = [label for sublist in true_labels_flat for label in sublist]
    y_pred_flat = [label for sublist in true_predictions_flat for label in sublist]
    
    cm = confusion_matrix(y_true_flat, y_pred_flat)
    error_distribution = np.diag(cm) / np.sum(cm, axis=1)

    plt.figure(figsize=(10, 6))
    colors = plt.cm.Oranges(np.linspace(0.3, 1, len(label_list))) 
    plt.bar(label_list, error_distribution, color=colors)
    plt.xlabel('Class', fontsize=14)
    plt.ylabel('Error Rate', fontsize=14)
    plt.title('Error Distribution by Class', fontsize=16)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()


plot_error_distribution(y_test_labels, y_pred_labels, label_list)
```

### Calculating Error Types

```python
def calculate_error_types(true_predictions_flat, true_labels_flat):
    # Flatten the lists of lists to get a single list of labels
    y_true_flat = [label for sublist in true_labels_flat for label in sublist]
    y_pred_flat = [label for sublist in true_predictions_flat for label in sublist]
    # Error types dictionary
    error_types = {
        'False Positive': {class_label: 0 for class_label in label_list},
        'False Negative': {class_label: 0 for class_label in label_list},
        'True Positive': {class_label: 0 for class_label in label_list},
        'True Negative': {class_label: 0 for class_label in label_list}
    }

    # Iterate over each prediction and true label pair
    for pred, true_label in zip(y_pred_flat, y_true_flat):
        # Iterate over each tag
        for class_label in label_list:
            if pred == class_label and true_label == class_label:
                error_types['True Positive'][class_label] += 1
            elif pred == class_label and true_label != class_label:
                error_types['False Positive'][class_label] += 1
            elif pred != class_label and true_label == class_label:
                error_types['False Negative'][class_label] += 1
            elif pred != class_label and true_label != class_label:
                error_types['True Negative'][class_label] += 1

    return error_types


def print_error_types(error_types):
    for error_type, class_counts in error_types.items():
        print(error_type)
        for class_label, count in class_counts.items():
            print(f"  {class_label}: {count}")


error_types = calculate_error_types(y_test_labels, y_pred_labels)
print_error_types(error_types)
```

```python
def visualize_error_types(error_types):
    # Extract error types
    false_positives = {k: v for k, v in error_types['False Positive'].items()}
    false_negatives = {k: v for k, v in error_types['False Negative'].items()}
    true_positives = {k: v for k, v in error_types['True Positive'].items()}
    true_negatives = {k: v for k, v in error_types['True Negative'].items()}

    # Get class labels
    class_labels = list(false_positives.keys())

    # Visualize error types with grouped bar chart
    fig, ax = plt.subplots(figsize=(10, 6))
    bar_width = 0.2
    index = np.arange(len(class_labels))

    # Plot bars for each error type
    bars1 = ax.bar(index, false_positives.values(), bar_width, label='False Positive', color='red')
    bars2 = ax.bar(index + bar_width, false_negatives.values(), bar_width, label='False Negative', color='blue')
    bars3 = ax.bar(index + 2 * bar_width, true_positives.values(), bar_width, label='True Positive', color='green')
    bars4 = ax.bar(index + 3 * bar_width, true_negatives.values(), bar_width, label='True Negative', color='orange')

    ax.set_xlabel('NER Tag', fontsize=16)
    ax.set_ylabel('Count', fontsize=16)
    ax.set_title('Distribution of Error Types', fontsize=16)
    ax.set_xticks(index + 1.5 * bar_width)
    ax.set_xticklabels(class_labels, rotation=45, ha='right')
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
visualize_error_types(error_types)
```
